```python
df = pd.read_pickle("202012-202103_processed.pkl")
```

```python
content=df.content_for_lda.to_list()
content=[x.lower() for x in content]
```

```python
remove_list = ['-pron-','not','say']
corpus = []
for index in range(len(content)):
    temp = content[index].split(" ")
    #temp = [x.lower() for x in temp]
    sen = [word for word in temp if not (len(word) < 3 or word in remove_list)]
    corpus.append(sen)
```

```python
model=LdaMulticore.load('models/23topics.model')
```

```python
top_topics = model.top_topics(texts=corpus, topn=20, coherence='c_v',dictionary=dictionary)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics

from pprint import pprint
pprint(top_topics)

print(num_topics,'Topics')
print('Average topic coherence: %.4f.' % avg_topic_coherence)
```

```python
import time
import numpy as np

doc_topic = []
t1=time.time()
print('...')
for i in range(len(bow)):
  temp=model[bow[i]] #
  if i % 10000 == 0:
      print('', i)
    
  p = [x[-1] for x in temp] #

  if len(p) == 23 and len(set(p)) == 1: #topic=-1
    #print(p)
    doc_topic.append(-1)
    
  else:
      topic_distribution = np.array(temp)
      #print(topic_distribution)
      line_index = topic_distribution.argmax(axis = 0) #
      topic = int(topic_distribution[line_index[-1],0])
      doc_topic.append(topic)
      
t2=time.time()
print('',t2-t1,'s')
```

```python
doc_topic = pd.DataFrame(doc_topic)
doc_topic.columns = ['Topic']
#doc_topic.index = [i for i in range(len(doc_topic))]
```
